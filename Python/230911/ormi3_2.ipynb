```python
# False => 문자열의 비교 연산은 0번 인덱스의 문자부터 유니코드를 비교하는데,
# 각 문자열의 4번 인덱스는 'o'와 ' '(공백)으로 각각의 유니코드는 0060F와 0020으로 'o'의 유니코드가 더 크기때문에
# False입니다.
print('hello' < 'hell o')
# False => 리스트의 비교 연산은 0번 인덱스부터 비교 연산을 진행하는데 각 리스트의 1번 인덱스는 20과 19로
# 왼쪽 리스트의 1번 인덱스가 더 크기 때문에 False입니다.
print([10, 20, 30] < [10, 19, 100])
# Fasle => 10 % 3.3은 부동소수점으로 인해 0.10000000000000053 값을 가지기 때문에 False 입니다.
# 만약 아래와 같은 연산의 목적으로 True를 얻기 위해서는 10과 3.3에 각각 10을 곱한 후
# % 연산을 진행하고 다시 10으로 나눈 후 0.1과 비교 연산을 진행해야 합니다.
print(10 % 3.3 == 0.1)
```
