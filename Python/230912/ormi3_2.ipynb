```python
# global을 하면 함수가 얼마나 중첩되어 있건 최상단에 있는 a를 바라봄
a = 10
def f():
    a = 100
    print(a)
    def ff():
        a = 1000
        print(a)
        def fff():
            global a
            print(a)
        fff()
    ff()
f()
print(a)
```

```python
# nonlocal
a = 10
def f():
    a = 100
    print(f'f a: {a}')
    def ff():
        a = 1000
        print(f'ff a: {a}')
        def fff():
            nonlocal a
            a = 100
            print(f'fff a: {a}')
        fff()
        print(f'ff a: {a}')
    ff()
f()
print(f'global a: {a}')
```

```python
# 상식 수준의 재귀
# 1. 팩토리얼
# 2. 문자열 뒤집기
```

```python
def factorial(n):
	if n <= 1:
		return 1
	return n * factorial(n-1)

factorial(5)
'''
함수호출				return
factorial(n)		n * factorial(n-1)
factorial(5)		5 * factorial(4) == 5 * 24
factorial(4)		4 * factorial(3) == 4 * 6
factorial(3)		3 * factorial(2) == 3 * 2
factorial(2)		2 * factorial(1) == 2 * 1
factorial(1)		1
'''
```

```python
def reversed_str(s):
	if len(s) <= 1:
		return s
	return reversed_str(s[1:]) + s[0]

reversed_str('hello')
# 함수호출							return
# reversed_str(s)				reversed_str(s[1:]) + s[0]
# reversed_str('hello')	reversed_str('ello') + 'h'	=> 'olleh'
# reversed_str('ello')	reversed_str('llo') + 'e'		=> 'olle'
# reversed_str('llo')		reversed_str('lo') + 'l'		=> 'oll'
# reversed_str('lo')		reversed_str('o') + 'l'			=> 'ol
# reversed_str('o')			'o'
```

```python
data = [
	[2, 'hello world', 30],
	[3, 'hworl', 33],
	[5, 'hello w', 31],
	[4, 'hello wo', 35],
	[6, 'hello w', 39],
]

# 1번 문제: 첫 번째 요소의 값으로 정렬
sorted(data)

# 2번 문제: 두 번째 요소의 문자열 길이로 정렬
sorted(data, key=lambda x : len(x[1]))

# 3번 문제: 세 번째 요소 값으로 정렬
sorted(data, key=lambda x : x[2])

# 4번 문제 : 첫 번째 요소와 세 번째 요소의 차가 큰 순으로 출력
sorted(data, key=lambda x : x[0] - x[2])
```
